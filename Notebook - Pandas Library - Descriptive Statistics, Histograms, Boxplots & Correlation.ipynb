# Aulas 12 e 13 - Boxplots & Correlações entre Variáveis

## O que são Boxplots?

- São gráficos de caixa que visam representar a variabilidade de uma variável aleatória. Apresentam medidas de posição e identificam outliers
- Nele são representados:
    - Min --> traço inferior
    - Q1 (25 %) --> base da caixa
    - Q2 (mediana) --> traço no meio da caixa
    - Q3 (75%) --> topo da caixa
    - Max --> traço superior
    - Outliers --> pontos além dos traços


## O que são Outliers?

- São valores discrepantes do restante dos dados, muito acima ou abaixo da média
- Algumas definições de outliers:
    - Valores distantes de 2 a 3 desvios padrões em relação à média
    - Valores distantes 1.5x do Q1 ou Q3

## df.boxplot ( column = 'col_name1' , by = 'col_name2' )

- column --> variável alvo que se deseja analisar a variabilidade e possíveis outliers
- by --> variável de agrupamento. Deve ser discreta/categórica (e.g. se quero plotar boxplots de preços de imóveis por número de quartos, a variável de agrupamento é o número de quartos)

## Correlação entre Variáveis

- Informa o nível de dependência entre duas variáveis

### df.corr (method = ' ' )

- Imprime uma matriz com todas as correlações LINEARES (Pearson) entre variáveis contidas no dataframe "df"
- Parâmetro "method":
    - 'pearson' (default) --> correlação LINEAR (paramétrico)
    - 'spearman' --> correlação RANKEADA (não paramétrico)
    - 'kendall' --> correlação TAU DE KENDALL (não paramétrico)
- Podemos adicionar **.plot( )** após o .corr ( ) para plotarmos gráficos de LINHAS de correlações

```
# Matriz de correlações LINEARES entre todas as variáveis do df
df.corr()
```

#### Coeficiente de Correlação de Spearman (RANK)

- É uma medida não paramétrica da correlação de postos (dependência estatística do ranking entre duas variáveis)
- Não se limita a identificar correlações lineares, como o Coeficiente de Correlação de Pearson
- Identifica correlações MONÓTONAS
- Uma função monótona é aquela que nunca diminui ou nunca aumenta, já que é um aumento variável independente:
    - Monotonicamente crescente: quando a variável “x” aumenta e a variável “y” nunca diminui
    - Diminui monotonicamente: quando a variável “x” aumenta, mas a variável “y” nunca aumenta
    - Não monótono: quando a variável “x” aumenta e a variável “y” as vezes aumenta e as vezes diminui

```
# Matriz de correlações RANKEADAS entre todas as variáveis do df
df.corr(method='spearman')
```

```
# Matriz de correlações TAU DE KENDALL entre todas as variáveis do df
df.corr(method='kendall')
```
