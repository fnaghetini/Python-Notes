## open ( )

- Função nativa do python para ler e editar documentos
- Ela cria um objeto do tipo ARQUIVO
- Parâmetro **mode =**
  - ' r ' -> Modo padrão de abertura. Permite apenas leitura do arquivo.
  - ' w ' -> Modo de escrita. Permite escrita no arquivo.
  - ' a ' -> Modo de escrita do tipo acrescentar. Permite acrescentar informação no final do arquivo.

```
arquivo = open('/content/drive/My Drive/Colab Notebooks/Data Science do ZERO/4 - Dominando o Python/Materiais Disponibilizados/automobile.csv',
     mode='r')
```

```
# Mostra que foi criado um objeto do tipo ARQUIVO
print(arquivo)
```

## file.readlines ( )

- Transforma o arquivo em um objeto do tipo LISTA
- Lê todas as linhas do arquivo como uma lista
- Cada elemento da lista gerada é uma linha do arquivo
- Portanto é muito útil para realizar iterações e é o método de arquivos MAIS USADO


```
lista = arquivo.readlines()
# Observe que o objeto virou uma LISTA
print (type(lista))

# OBS.: não chamei a variável 'lista', pois ela é muito grande!
```

```
# Loop para imprimir cada elemento (linha do arquivo) da lista
for i in lista:
  print (i)
```

## file.tell ( )

- Retorna a quantidade de caracteres que existem em um arquivo

```
# O arquivo apresenta 26139 caracteres
arquivo.tell()
```

## file.read ( )

- Escaneia e imprime todo o arquivo
- Para visualizar o resultado, devemos imprimir

```
print(arquivo.read())
```

## Iterações sobre arquivos

- Podemos iterar diretamente sobre ARQUIVOS, pois o python reconhece esse tipo de objeto como SEQUENCIAL
- O método readlines leva vantagem, pois é mais fácil iterar sobre listas do que sobre objetos
- Entretanto, esse tipo de iteração é MAIS LEVE do que uma iteração sobre uma lista, ou seja, ocupa MENOS MEMÓRIA

```
arquivo = open('/content/drive/My Drive/Colab Notebooks/Data Science do ZERO/4 - Dominando o Python/Materiais Disponibilizados/automobile.csv',
     mode='r')
```

```
for linha in arquivo:
    print (linha)
```

# Aula 18 - Escrevendo em Arquivos

## open ('file_name.txt', mode = 'w')

- Permite abrir um objeto como ARQUIVO e escrever nele

```
# Abre o arquivo em modo de escrita. 'resultados.txt' não existia
file2 = open('resultados.txt','w')
```

## file.write('o_que_vc_quer_escrever')

- Esse método é usado para para escrever em arquivos no python
- As informações escritas no arquivo via este método ainda NÃO PERSISTEM no disco (precisamos usar o método 'close ( )' para o fazer

```
# Retorna um código e escreve no arquivo
file2.write('Eu me tornarei um data scientist')
file2.write('Data Science é a profissão do futuro!')
```

## file.close ( )

- Método 'fecha' o arquivo, ou seja, salva as informações adicionadas ao arquivo no disco rígido do PC (persiste as informações no disco)

```
file2.close()
```

## file.read ( )

- Não é possível ler o arquivo após ele ter sido 'fechado' pelo método 'close ( )'
- Não é possível ler o arquivo com ele aberto no modo escrita, ou seja, "file.open('nome_arquivo', mode = 'w')"
- Devemos usar o método "file.open('nome_arquivo', mode = 'r')" e, em seguida, usar algo para lê-lo, como, por exemplo o método "file.readlines( )"

```
file2 = open('resultados.txt', mode='r')
```

```
file2.readlines()
```

## open ('file_name.txt', mode = 'a')

- Permite abrir um objeto como ARQUIVO e escrever ao final desse arquivo
- Método que ACRESCENTA informações ao arquivo
- Caso eu abrisse o arquivo no modo 'w' (de escrita) para adicionar novas informações, o python, na verdade, sobrescreveria toda a informação prévia. Daí surge a necessidade do modo 'a' (acrescentar informações)

```
file2 = open('resultados.txt', mode='a')
```

```
# Adicionando mais informação ao arquivo
file2.write('Última linha do arquivo.')
```

```
# Fechando o arquivo
file2.close()
```

```
# Lendo novamente o arquivo no modo 'r'
file2 = open('resultados.txt',mode='r')
```

```
# Lendo o arquivo como uma lista
file2.readlines()
```

# Aula 19 - Outros Métodos para Arquivos

## file.writelines ( )

- Método que permite escrever uma lista em um arquivo

```
# Criando uma lista de strings
lista = ["Manipular arquivos no Python é fácil","Segunda Linha!","Mais uma linha!"]
```

```
# Abre o arquivo no modo escrita (isso sobrecreverá todas as infos do arquivo)
file2 = open('resultados.txt','w')
```

```
# Escreve a lista de strings
file2.writelines(lista)
# 'Fecha' (persiste) as informações no arquivo
file2.close()
```

```
# Abre para leitura e imprime no formato de lista
file2 = open('resultados.txt','r')
file2.readlines()
```

## Trabalhando com a Instrução WITH

- Com essa instrução WITH, não é preciso fechar o arquivo explicitamente
- Essa instrução fecha o arquivo quando os blocos dele não estão sendo mais utilizados
- A vantagem é que, se esquecermos de usar a instrução "close" no código, o Python fecha o arquivo automaticamente após uma operação

```
# O arquivo é lido e seu conteúdo é atribuído à variável "file2". Após isso (a 
# execução de todo o bloco), o arquivo é fechado automaticamente. Não precisamos
# usar a instrução 'close' mais
with open("resultados.txt") as file2:
    print (file2.read())
```

```
# Vamos validar pra ver se o arquivo foi, de fato, fechado sem o uso do 'close'

#file2.read()

# Deixei comentado, pq realmente o arquivo foi fechado
```

```
# Cria uma lista com as linhas do arquivo e fecha apos concluir
with open("/content/drive/My Drive/Colab Notebooks/Data Science do ZERO/4 - Dominando o Python/Materiais Disponibilizados/automobile.csv") as file2:
    data = file2.readlines()
```

```
# Arquivo já está fechado!
#file2.read()
```

```
# Imprime uma única lista com as linhas do arquivo 
data
```

```
# Verificação do tipo do arquivo. De fato é uma lista!
type(data)
```

## linha.split ( )

- Podemos usar o método 'split ( )' para quebrar cada linha em algum caracter especificado
- Como não foi especificado nenhum caracter, por padrão, a quebra ocorrerá nos espaços
- Repare que cada linha da lista virou um lista. Temos, portanto, uma lista aninhada

```
for linha in data:
    print (linha.split())
```
