# Aula 05 - Manipulação de Listas

## Listas [ ]

- Conjunto de objetos;
- Podem ser aninhadas (listas dentro de listas - matrizes);
- Podem apresentar variáveis/objetos de tipos diferentes;
- Diferentemente das **strings** as listas são **mutáveis**, ou seja, é possível modificar seus elementos (não é possível modificar caracteres de strings)

```
# Sintaxe geral das listas
bichos = [1,2,3,4,5]
```

```
# Listas aninhadas
anin = [[1,2,3],[4,5,6],[7,8,9]]
```

```
# Listas mistas
mista = [1,2.5,"Cavalo",(2,3),[1,4,5],True]
```

## Busca e Modificação por Índice de Listas

```
# Primeiro termo da lista tem índice 0
bichos[0]
```

```
# O terceiro elemento é o de índice 2 e é uma lista
anin[2]
```

```
# Modificação de um elemento
anin[2] = "Burro"
anin
```

### Slicing em Listas

```
numeros = [11,22,33,44,55,66,77,88,99,56,76,34,21,23,34,45]
# Busca por todos os elementos da lista a partir do índice 1 (segundo elemento)
numeros[1:]
```

```
# Busca entre os índices 2 e 10
numeros[2:10]
# Note que o número de elementos de saída é igual a (índice final - índice inicial)
```

```
# Busca pelo último elemento
numeros[-1]
```

```
# Busca do antepenúltimo elemento até o último
numeros[-3:]
# Note que não se coloca o último elemento, pois queremos 3 elementos e não 2
```

```
# Busca do primeiro ao 10 elemento
numeros[:10] # resultados output = 10 - 0 = 10
```

### lista.remove( ' ' )

- Remove o elemento indicado da lista

```
nomes = ["Maria","Marcos","Camila","Franco","Carlindo","Letícia"]
nomes.remove("Franco")
nomes
# Ao contrário do R, aqui a operação de remoção de um elemento é permanente e
# não precisa de atualizar a variável
```

## Alguns Métodos e Operações de Listas

### len(lista)

- Retorna o número de elementos contidos em uma lista
- Método muito performático;
- Muito poderoso quando aliado ao laço "for"


```
# Número de elementos da lista "numeros"
len(numeros)
```

```
len(anin)
```

```
len(bichos)
```

### Concatenação de Listas

- Serve para unir duas listas
- A concatenação de listas NÃO É PERMANENTE e exige declaração de variável para ser armazenada
- A ordem de concatenação importa!

```
bichos + ['Cavalo','Vaca','Boi','Avestruz']
```

```
# Note que a concatenação não foi definitiva
bichos
```

```
# Note que a ordem de concatenação altera o resultado
['Cavalo','Vaca','Boi','Avestruz'] + bichos
```

```
# Salvando a concatenação em uma nova variável
bichos = ['Cavalo','Vaca','Boi','Avestruz'] + bichos
```

### elemento **in** lista

- Útil para verificar se existe um determinado elemento em uma lista
- Retorna um valor booleano (T/F)

```
# Existe o elemento 'Cavalo' na minha lista 'bichos'?
'Cavalo' in bichos
```

```
# Não existe o elemento '1' na lista 'anin'
1 in anin
```

### max(lista)

- Retorna o elemento de valor máximo de uma lista
- Não pode ser usado em listas que contém strings e inteiros simultaneamente
- Se a lista apresentar floats e ints, é possível usar essa função
- Retorna o último elemento de uma ordenação alfabética em listas constituídas apenas por strings

```
# Valor máximo de uma lista com ints e floats
lista1 = [1,2.0,4,5,6,7]
max(lista1)
```

```
# Quando aplicado a uma lista de strings, retorna o último elemento de uma
# ordenação alfabética
max(nomes)
```

### min(lista)

- Retorna o elemento de valor mínimo de uma lista
- Não pode ser usado em listas que contém strings e inteiros simultaneamente
- Se a lista apresentar floats e ints, é possível usar essa função
- Retorna o primeiro elemento de uma ordenação alfabética em listas constituídas apenas por strings

```
min(lista1)
```

```
min(nomes)
```

### Multiplicação de elementos de uma lista

```
# Multiplicação de uma lista de string por uma constante
['Olá'] * 5
```

```
# Multiplicação de uma lista de ints por uma constante resulta numa replicação
# da de cada elemento n vezes, sendo constante = n
dec = [1,2,3]
dec * 2
```

# Aulas 8 e 9 - Introdução a Tuplas

- São similares às listas, mas as tuplas, ao contrário das listas, são IMUTÁVEIS, ou seja, não podem ter seus elementos alterados;
- Há uma **diferença conceitual** também entre listas e tuplas:
  - Listas -> apresentam uma sequência de elementos HOMOGÊNEOS, do mesmo tipo;
  - Tuplas -> apresentam uma sequência de elementos HETEROGÊNEOS. de tipos diferentes;
- Tuplas são usadas, por exemplo, para armazenar coordenadas geográficas lat-long. Isso pq tratam-se de elementos de significados distintos e pq não podem ser alterados
- Tuplas podemos apresentar objetos de tipos diferentes
- Podemos criar tuplas ANINHADAS
- A tupla é utilizada para declarar elementos que não podem ser alterados de jeito nenhum

```
# Sintaxe geral de uma tupla
tupla_geral = ()
# Foi criada uma tupla vazia
```

```
# Tupla mista
tup = (1,2.5,"Camila","Henrique",True)
tup
```

## Busca de elementos da tupla pelo índice

- Igual à busca por índice de listas
- Podemos buscar mais de um elemento pelo índice por linha de código. Se isso é feito com uma lista, o resultado devolvido é uma tupla

```
# Retorna o terceiro elemento da tupla
tup[2]
```

```
# Podemos imprimir vários elementos ao mesmo tempo
tup[0],tup[2],tup[3]
# Retorna UMA tupla com 3 elementos
```

```
# Se fizermos o mesmo para uma lista, o resultado devolvido é uma tupla
numeros[1],numeros[6],numeros[10]
```

```
# Alterar um elemento da tupla não é possível, pois é um tipo de objeto IMUTÁVEL
#     tup[1] = 5
```

## Tuplas Aninhadas

- Podemos fazer a busca de índice ' [ ] [ ] ' igual fazemos para as listas

```
# As tuplas podem ser aninhadas
tup_a = (3,4,(2,3,4),(95,6,4))
tup_a
```

```
# Podemos buscar elementos dentro de um elemento aninhado de uma tupla igual
# fazemos para listas
tup_a[2][2]
```

```
# Outro exemplo da situação acima
tup_a[3][1]
```

## Operações Básicas com Tuplas

### Multiplicação de uma tupla por uma constante 'n'

- Retorna uma multiplicação dos elementos da tupla pela constante (assim como ocorre para as listas)
- Esse tipo de operação NÃO é PERMANENTE e deve ser salva em uma variável (assim como ocorre para as listas)

```
tup_a * 3
```

### Concatenação de Tuplas

- Igual à concatenação de listas

```
(1,2,3,4) + (2,3,4,5)
```

### elemento **in** tupla

- Método que verifica se existe um determinado elemento em uma tupla (assim como para listas)
- Retorna um valor booleano (T/F)

```
3 in tup_a
```

## Slicing de Tuplas

- Funciona do mesmo jeito que para as listas

```
tup_a[:2]
```

```
tup_a[-1]
```

```
tup_a[-3:]
```

```
tup[1:3]
```

## Métodos de Tuplas

- Existem menos métodos aplicados a tuplas quando comparado aos métodos disponíveis para listas

### tupla.count(elemento)

- Retorna quantas vezes determinado elemento aparece na tupla (igual a esse método para listas)

```
coisas = ('Martelo','Canivete','Canivete','Maçarico','Cigarro')
coisas.count('Canivete')
```

### tupla.index(elemento)

- Retorna o índice referente ao elemento da tupla definido (igual a esse método para listas)

```
coisas.index('Cigarro')
```

```
# Nesse caso, como existem dois elementos com o mesmo nome, este método retorna
# o índice da primeira aparição desse elemento
coisas.index('Canivete')
```

### len(tupla)

- Retorna o número de elementos contidos em uma tupla (igual e este método para listas)
- De novo, muito poderoso quando aliado ao laço "for"

```
len(coisas)
```

### max(tupla)

- Retorna o maior valor de uma tupla
- Não pode ser usado em tuplas que contenham ints e strings simultaneamente
- Pode ser usado em tuplas que contenham floats e ints
- Caso a tupla seja constituída apenas por strings, retorna o último elemento da ordenação alfabética

```
max(coisas)
```

```
numerais = (1,22,33,45,56,76,78,7,8,98,9)
max(numerais)
```

### min(tupla)

- Retorna o menor valor de uma tupla
- Não pode ser usado em tuplas que contenham ints e strings simultaneamente
- Pode ser usado em tuplas que contenham floats e ints
- Caso a tupla seja constituída apenas por strings, retorna o primeiro elemento da ordenação alfabética

```
min(coisas)
```

```
min(numerais)
```

# Aula 10 - Trabalhando com Dicionários

- Funcionam na estrutura CHAVE : VALOR
- Assim como as listas, os dicionários são MUTÁVEIS (podem ser alterados após a sua criação)
- Podem conter dados de diferentes tipos e, inclusive, outros dicionários

## Dicionários { }

- Pode conter, nos valores, listas, strings, ints, floats e até mesmo outros dicionários

```
# Sintaxe geral. Criação de um dicionário vazio
dic = {}
# O dicionário é do tipo "dict"
type(dic)
```

```
# Criação de um dicionário bem mais complexo. Repare que existe um dicionário 
# dentro do dicionário criado. Observe a estrutura "chave : valor":
pessoas = {'Rodrigo':30, 'Marcela':30, 'Maria':55, 'Jose':80, 
           'pesos':{'Rodrigo':68,'Maria':70,'Jose':87},
           'Valores':[1,3.5,400,5,6], 
           'Strings': ['MinerandoDados.com.br','Data Science','Machine Learning']
           }
pessoas

# Chaves -> Rodrigo, Marcela, Maria, José, Pesos, Valores e Strings
# Valores -> tudo aquilo contido dentro das chaves
```

## Acesso a valores em um dicionário

- Devemos informar a chave para obter como retorno o(s) valor(es) associado(s) a essa chave

```
pessoas['Jose']
```

```
pessoas['pesos']
```

```
pessoas['Strings']
```

## Acesso a valores de um dicionário aninhado

- Funciona da mesma forma que a busca por listas e tuplas aninhadas. Mas, ao invés de índices, devemos informar duas chaves

```
pessoas ['pesos']['Maria']
```

```
pessoas['pesos']['Rodrigo']
```

## Acesso a valores de uma lista contida em um dicionário

```
pessoas['Strings'][1]
```

```
pessoas['Valores'][2]
```

```
pessoas['Strings'][0]
```

- dic ["key"] = valor

```
pessoas['Rodrigo'] = 99
pessoas
```

```
pessoas['Jose'] = 10
pessoas['Jose']
```

## Adição de nova chave + valor(es) a um dicionário

- dic["new_key"] = valor

```
pessoas['Ricardo'] = 50
pessoas
```

```
pessoas['Joao'] = 100
pessoas
```

## Remoção de chave e valor(es) de um  - del

- del dic["key"]
- Este método é PERMANENTE (ao contrário do R)

```
del pessoas['Joao']
pessoas
```

```
del pessoas['Ricardo']
pessoas
```

# Aula 11 - Manipulação e Métodos de Dicionários

## dic.keys ( )

- Retorna as chaves do dicionário

```
pessoas.keys()
```

## dic.values ( )

- Retorna todos os valores do dicionário

```
pessoas.values()
```

## dic.update (new_dic)

- Atualiza um dicionário antigo com as chaves e valores de um novo dicionário criado

```
# Criando novo dicionário
pessoas2 = {'Henrique' : 66, 'Clara' : 89, 'Joaquim' : 88}
pessoas2
```

```
# Atualização do dicionário com as informações trazidas pelo novo dicionário
pessoas.update(pessoas2)
pessoas
```

## dic.get ("key","mensagem")

- Retorna o valor da chave definida. Caso a chave não exista, retorna uma mensagem
- Esse é muito importante quando queremos descobrir se existe uma determinada chave em um dicionário. Se não existir, podemos customizar para retornar um valor padrão

```
pessoas.get('Lucas','Chave inexistente.')
```

```
pessoas.get('Maria','Chave inexistente.')
```

## dic.setdefault ('key',valor)

- Retorna o valor relativo à chave definida. Se essa chave não existir, ele criar uma chave 'key' com o valor 'valor'

```
# Como a chave 'Lucas' não existia, ela foi adicionada ao dicionário juntamente com seu valor '31'
pessoas.setdefault('Lucas',31)
pessoas
```

## dic.pop("key","mensagem")

- Remove a chave e retorna o seu valor. Caso a chave não exista, retorna uma mensagem (segundo parâmetro)
- Permite a manipulação do dicionário de forma automatizada
- Muito útil quando usado em conjunto com laços

```
# A chave não existe
pessoas.pop('Marcelo','Essa chave não existe.')
```

```
# A chave foi apagada e seu valor foi devolvido
pessoas.pop('Rodrigo','Essa chave não existe.')
```

## dic.clear ( )

- Apaga todos os valores de um dicionário

```
# Apagando dicionário
pessoas.clear()
```

```
# Dicionário vazio
pessoas
```
