## IFs Aninhados

- Estratégia muito útil
- Deve ser usada com moderação, pois pode deixar o código muito compexo

```
usuario = input("Insira o usuário: ")
senha = input("Insira a senha: ")

if usuario == 'Franco':
  if senha == '12345':
    print('OK! Senha e usuário corretos.')
  else:
    print('Senha inválida. Gostaria de recuperar sua senha?')
else:
  print('Usuário incorreto!!!')
```

# Aulas 14 e 15 - Laços de Repetição

## FOR em listas

```
# Como não declarei a lista como uma variável, devemos escrever com essa sintaxe
# bizarra
for i in [3,6,4,7,9]:
  print ('valor: %s' %i)
```

```
objetos = ['cadeira','mesa','estojo','borracha']
for i in objetos:
  print ('Eu tenho um(a) {x}'.format(x=i))
```

## FOR em strings

```
for caracter in 'Eu amo Data Science!':
  print(caracter, end=' ')
```

## FOR com dicionários

```
# Imprime cada chave do dicionário
idades = {'Franco':23,'Camila':23,'Maria':62,'Marcos':63}
for i in idades:
  print (i)
```

## FOR com zip

- Itera sobre duas listas ao mesmo tempo
- Interessante para iterarmos sobre dois arquivos de texto simultaneamente

```
numeros = [1,4,5,6,7]
indices = ['a','b','c','d','e']
for i,j in zip(indices,numeros):
  print(i,j)
```

## FORs Aninhados

- Da mesma forma que todo comando aninhado, deve ser usado com cadência para não deixar o programa muito complexo

```
for i in ['1a Fase','2a Fase','3a Fase']:
    print (i)
    for j in ['Maça','Bananas','Uvas']:
        print (j)
```

## Instrução BREAK em FOR

- Utilizada para ENCERRAR um loop

```
for letra in 'Eu amo Python':
  if letra == 'o':
    break
  print ('letra = {x}'.format(x=letra))
```

## Instrução CONTINUE em FOR

- Utilizada para CONTINUAR um loop

```
# Repare que quando chega no 'o', o loop reinicia e não imprime 'o'
for letra in 'Eu amo Python':
  if letra == 'o':
    continue
  print ('letra = {x}'.format(x=letra))
```

# Aula 16 - Loops com WHILE

## WHILE simples

```
i = 0
while i < 10:
  print (i,end=', ')
  i = i + 1
```

## WHILE com condição ELSE

- O comando associado ao ELSE é executado quando o condição do WHILE passa a ser falsa

```
i = 0
while i < 10:
  print(i,end=', ')
  i = i + 1
else:
  print ('O número é menor ou igual a 10.')
```

## WHILE com instrução BREAK

```
i = 0
while i < 10:
  if i == 5:
    break
  print (i)
  i = i + 1
```

## Observações finais

- Pode-se usar WHILEs aninhados (com moderação)
- Pode-se usar WHILE com instrução CONTINUE
- Cuidado com os loops infinitos do WHILE!!!
