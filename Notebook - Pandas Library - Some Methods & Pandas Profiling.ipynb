# Aula 02 - Conhecendo Dataframes

## O que são Dataframes?

- São "tabelas" que apresentam linhas e colunas
- As colunas são chamadas de atributos/variáveis/features
- As linhas podem conter objetos de tipos distintos
- Dataframes aceitam qqr tipo de objeto

## df = pd.read_csv (path, sep = ' , ', header = i, index_col = 'col', usecols = [ 'col1' , 'col2' , 'col3' )

- Retorna um dataframe "df" a partir de um arquivo csv "path"
- sep =  --> serve para explicitar qual separador de elementos é utilizado no arquivo CSV
- index_col = 'col' --> define qual será o atributo que, por padrão, será será responsável pela indexação do df
- usecols = [ ] --> define quais atributos do arquivo csv devem ser importados

## Campos do relatório provido pelo 'Pandas Profiling'

- **Overview** --> mostra as principais características do dataset, incluindo linhas/colunas duplicadas, % de valores faltantes, número de variáveis
- **Variables** --> mostra as principais estatísticas descritivas de cada variável. Pode-se clicar em "Toggle Details" para analisar estatísticas mais específicas
- **Interactions** --> mostra as principais interações entre as variáveis, via scatterplot. Os scatterplots são interativos
- **Correlations** --> mostra os heatmaps com todos os métodos de correlação disponíveis
- **Missing Values** --> mostra as estatísticas relacionadas aos valores faltantes (relativa e absoluta)
- **Sample** --> mostra as 10 primeiras linhas do df
- **Warnings** --> mostra as advertências de eventuais características problemáticas no dataset (elevado números de valores NaN ou nulos, correlações elevadas) para auxiliar na tomada de decisões para manipular o dataset
