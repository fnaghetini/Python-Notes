## binom.cdf (num_success , num_trials , prob)

- Retorna a probabilidade CUMULATIVA binomial
- Parâmetros:
  - num_success --> número de sucessos desejado
  - num_trials --> número de experimentos
  - prob --> probabilidade associada ao evento

```
# Passaremos por 4 sinais de 4 tempos. Qual é a probabilidade cumulativa de não 
# pegar nenhum sinal verde?
binom.cdf(0,4,0.25)

# P(zero sinal verde) = Pacum(zero sinal verde)
```

```
# Passaremos por 4 sinais de 4 tempos. Qual é a probabilidade cumulativa de não 
# pegar 1 sinal verde?
binom.cdf(1,4,0.25)

# Note que: P(zero sinal verde) + P(1 sinal verde) = Pacum(1 sinal verde)
```

```
# Passaremos por 4 sinais de 4 tempos. Qual é a probabilidade cumulativa de não 
# pegar 2 sinais verdes?
binom.cdf(2,4,0.25)

# Note que: P(zero sinal verde) + P(1 sinal verde) + P(2 sinal verde) =
# = Pacum(2 sinal verde)
```

```
# Passaremos por 4 sinais de 4 tempos. Qual é a probabilidade cumulativa de não 
# pegar 4 sinais verdes?
binom.cdf(4,4,0.25)

# Note que: P(zero sinal verde) + P(1 sinal verde) + P(2 sinal verde) + 
# + P(3 sinal verde) + P(4 sinal verde) = Pacum(4 sinal verde) = 1
```

# Distribuição Normal

## Importação de Bibliotecas

```
from scipy.stats import norm
```

## norm.pdf (x,mean,std)

- Calcula probabilidade CUMULATIVA do tipo P(X <= x)
- Parâmetros:
  - x --> valor que deseja-se calcular a probabilidade do tipo P(X <= x)
  - mean --> média
  - std --> DP

```
# Há diversos objetos dentro de uma cesta. Sabe-se que a média é igual a 8 e o
# DP igual a 2. Qual é a probabilidade de se tirar um objeto de peso MAIOR que
# 6kg? Ou seja, calcule P(peso <= 6)
norm.cdf(6,8,2)
```

## norm.sf (x,mean,std)

- Calcula probabilidade CUMULATIVA do tipo P(X > x)
- Parâmetros:
  - x --> valor que deseja-se calcular a probabilidade do tipo P(X > x)
  - mean --> média
  - std --> DP

```
# Há diversos objetos dentro de uma cesta. Sabe-se que a média é igual a 8 e o
# DP igual a 2. Qual é a probabilidade de se tirar um objeto de peso MENOR que
# 6kg? Ou seja, calcule P(peso > 6)
norm.sf(6,8,2)
```

```
# Note que P(X > x) = 1 - P(X <= x)
norm.sf(6,8,2) == 1 - norm.cdf(6,8,2)
```

## Outros Problemas

```
# Há diversos objetos dentro de uma cesta. Sabe-se que a média é igual a 8 e o
# DP igual a 2. Qual é a probabilidade de se tirar um objeto de peso MENOR que
# 6kg ou MAIOR que 10kg? Ou seja, calcule P(peso <= 6) + P(peso > 10). Note que
# os eventos são mutuamente excludentes
norm.cdf(6,8,2) + norm.sf(10,8,2)
```

```
# Há diversos objetos dentro de uma cesta. Sabe-se que a média é igual a 8 e o
# DP igual a 2. Qual é a probabilidade de se tirar um objeto de peso MENOR que
# 10kg e MAIOR que 8kg? Ou seja, calcule P(peso <= 10) - P(peso > 8).
norm.cdf(10,8,2) - norm.sf(8,8,2)
```

## Verificação da Normalidade de uma Distribuição

### Importação de Bibliotecas

```
from scipy import stats
from scipy.stats import norm
import matplotlib.pyplot as plt
```

### norm.rvs(size = n)

- Gera números aleatórios normalmente distribuídos (Distribuição Normal Padrão, i.e. média ~ 0 e DP ~ 1)
- Parâmetros:
  - size = n --> número de números aleatórios

```
normal = norm.rvs(size=100)
```

### stats.probplot(data , plot = plt)

- Plota um Q-Q Plot que verifica se a distribuição de um conjunto de dados é de fato normal
- Parâmetros:
  - data --> conjunto de dados que deseja-se verificar normalidade
  - plot = plt --> renderiza o Q-Q Plot

```
stats.probplot(normal,plot=plt)
```

### stats.shapiro(data)

- Realiza o Teste de Shapiro-Wilk para verificar a normalidade de um conjunto de dados, utilizando um teste de hipótese, cuja hipótese nula (H0) é a normalidade da distribuição
- Parâmetros:
  - data --> conjunto de dados que deseja-se verificar normalidade
- Outputs:
  - (estatística de teste, valor-p)
- Resultados:
  - statistic_test < p-value --> rejeita H0 (dados não normalmente distribuídos)
  - statistic_test > p-value --> não rejeita H0 (dados normalmente distribuídos)

```
stats.shapiro(normal)
```

## t.cdf (t_stat,df)

- Calcula a probabilidade ACUMULADA do tipo P(X <= x)
- Parâmetros:
  - t_stat --> estatística "t" encontrada pela fórmula acima
  - df --> número de graus de liberdade (n - 1)

### Exemplo I - Probabilidade MENOR que (X < x)

Uma pesquisa mostra que cientistas de dados ganham 75.00 reais por hora. Uma amostra de 9 cientistas de dados é selecionada para avaliar o preço da hora de cada um. O DP da amostra é igual a 10.

Qual é a probabilidade do salário de um cientista de dados ser MENOR que 80.00 reais por hora?

OBS.: Não se conhece o DP da população é n < 30. Portanto, deve-se utilizar a Distribuição T-Student

- Sumarizando as informações:
  - Média (população) = 75
  - DP (amostra) = 10
  - Tamanho amostra (n) = 9
  - **Graus de liberdade (t) = 8**

- Cálculo do parâmetro "t" pela fórmula:
  - **t = 1.5** .

```
t.cdf(1.5,8)
```
